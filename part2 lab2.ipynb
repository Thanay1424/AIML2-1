```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    'people': ['kiran', 'arun', 'vijay', 'varun'],
    'age': [25, 30, 35, 40],
    'height': [145, 151, 165, 173],
    'weight': [45, 55, 65, 75]
}

df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['height'], color='blue', s=100, alpha=0.7, edgecolors='black')
plt.title('Scatter Plot of Age vs. Height')
plt.xlabel('Age (years)')
plt.ylabel('Height (cm)')
plt.grid(True)

x = np.arange(len(df['people']))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 7))
bars1 = ax.bar(x - width/2, df['height'], width, label='Height (cm)')
bars2 = ax.bar(x + width/2, df['weight'], width, label='Weight (kg)')

ax.set_xlabel('People')
ax.set_ylabel('Height and Weight')
ax.set_title('Bar Chart of Height and Weight for Each Person')
ax.set_xticks(x)
ax.set_xticklabels(df['people'])
ax.legend()

plt.figure(figsize=(8, 6))

plt.subplot(1, 3, 1)
plt.hist(df['age'], bins=4, color='green', edgecolor='black')
plt.title('Histogram of Age Distribution')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.show()

plt.subplot(1, 3, 2)
plt.hist(df['height'], bins=4, color='orange', edgecolor='black')
plt.title('Histogram of Height')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(df['weight'], bins=5, color='orange', edgecolor='black')
plt.title('Histogram of Weight')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
